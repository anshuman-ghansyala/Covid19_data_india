# Covid19 in India#

Coronavirus disease 2019 (COVID-19) is a contagious disease caused by severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The first case was identified in Wuhan, China, in December 2019. The disease has since spread worldwide, leading to an ongoing pandemic.

For this project i will be focusing only on the Covid19 situation in India. 
The first case of COVID-19 in India was reported on 30 January 2020 and the cases have been increasing ever since. 


Firstly let us start with importing necessary libraries.
I have imported pandas and matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

I have imported the necessary data from kaggle, this data has been provided on kaggle by Sudalai Rajkumar (https://www.kaggle.com/sudalairajkumar/covid19-in-india).

```python
covid19_data = pd.read_csv(r"E:\Anshuman\covid_19_data.csv", index_col = 0)
```

Let us take a look at the data, for this we use head and tail function of pandas. 

```python
print(covid19_data.head())
print(covid19_data.tail())
```

In the original data we have 7 columns namely Serial number, Date, Time, State, Cured, Deaths and Confirmed. For my convenience I have added another column named Month which record the name of the Month in which the data was recorded.
We notice that we have data from 30th January 2020 till 9th December 2020.

Now let us broadly look at the maximum value of individuals Cured, dead and confirmed. We notice that the maximum values in the columns named Cured, Deaths and Confirmed were 
173708, 47827, 1859367 respectively. 


```python
print(covid19_data[['Cured','Deaths','Confirmed']].max())
```

From analysing the data we know that we have the latest data of the date 9th December 2020. Let us now select data for this particular date from our original data and sort this data by number of confirmed individuals in descending order. 

```python
last_date = covid19_data[covid19_data['Date'] == '09-12-2020']
sorted_last_date = last_date.sort_values(by = 'Confirmed',ascending = False)
print(sorted_last_date.head())
```

We would like to plot a bar graph showing us the top five states with maximum number of confirmed individuals. We select the top five states from our data frame and plot a bar graph.

```python
xaxis = sorted_last_date.iloc[0:5,3]
print(xaxis)
yaxis = sorted_last_date.iloc[0:5,6]
print(yaxis)
```

We notice that Maharashtra has approximately 18.5 Lakhs number of confirmed individuals the maximum among all the states followed by Karnataka, Andhra Pradesh, Tamil Nadu and Kerala.  

```python
fig = plt.figure(figsize = (10,15))
plt.bar(xaxis,yaxis, color = 'r')
values = ['18.5 lakhs','8.9 lakhs', '8.7 lakhs', '7.9 lakhs', '6.4 lakhs']
plt.yticks(yaxis,values)
plt.xlabel('States')
plt.ylabel('No of individual confirmed')
plt.title('States with maximum Confirmed Cases')
plt.xticks(rotation = 45)
plt.show()
```

We would also like to plot a bar graph depicting the states with the least number of confirmed individuals. We select the necessary data from the data frame and plot it as a bar graph. 

```python
print(sorted_last_date.iloc[-6:-1,:])
```

```python
x1axis = sorted_last_date.iloc[-6:-1,3]
y1axis = sorted_last_date.iloc[-6:-1,6]
print(x1axis)
print(y1axis)
```

We Notice that Mizoram has approximately 4000 number of confirmed individuals which is least among all the states followed by Andaman and Nicobar Islands, Sikkim, Ladakh and Nagaland.

It is only fair to compare the number of confirmed cases as a percentage of population. To do so let us access a excel file containing the population of every state and union territories.

```python
population_data = pd.read_excel("E:\Anshuman\Population_data19.xlsx")
print(population_data.head())
```

Let us merge this data with our earlier data using the merge function of pandas with an inner join. 

```python
merged_data = sorted_last_date.merge(population_data, on = 'State' )
print(merged_data.head())
```

Let us crete a new column 'Confirmed per population' which would be the ratio of number of confirmed cases in a particular state by their respective population. 

```python
merged_data['Confirmed per population'] = (merged_data['Confirmed']/merged_data['Population in 2020'])*100
print(merged_data.head())
```

Now let us see which states have the highest number of confirmed cases per population. We sort our data on basis of the column 'Confirmed per population' in descending order.

```python
sorted_merged_data = merged_data.sort_values( by = 'Confirmed per population', ascending = False)
print(sorted_merged_data.head())
```

We notice that although puducherry and Mizoram have very low number of confirmed individual , they are the top two states when it comes to number of Confirmed cases by population.
Now let us plot a bar garph to show the Highest number of confirmed individuals by poplation. 

```python
new_xlabel = sorted_merged_data.iloc[0:5,3].to_list()
print(new_xlabel)
new_ylabel = sorted_merged_data.iloc[0:5,8].to_list()
print(new_ylabel)
```

We notice that Puducherry has approximately 4.5% of confirmed individuals which is the highest percentage of Confirmed individual among all states and Union Territories followed by Mizoram, Delhi, Kerala and Andhra Pradesh.

```python
fig = plt.figure()
plt.bar(new_xlabel, new_ylabel, color = 'r')
plt.xlabel('State')
plt.ylabel('Confirmed per population')
plt.title('No of Confirmed Cases per population')
plt.show()
```

Similarly we plot a bar graph for lowest number of confirmed cases per population.

```python
new_x1label = sorted_merged_data.iloc[-6:-1,3].to_list()
print(new_x1label)
new_y1label = sorted_merged_data.iloc[-6:-1,8].to_list()
print(new_y1label)
```

we notice that Uttar pradesh has aprroximately 0.23 % number of confirmed individuals by population the least among all the states and union territories followed by Madhya Pradesh, Jharkhand, Gujarat and Meghalaya.

Now let us look at which states have higher number of cured and death cases as a percentage of number of Confirmed individuals. 

```python
merged_data['Cured %'] = (merged_data['Cured']/merged_data['Confirmed'])*100
merged_data['Deaths %'] = (merged_data['Deaths']/merged_data['Confirmed'])*100
print(merged_data.head())
```

We notice that the Cured as percentage of Confirmed is highest among Andhra Pradesh and lowest in Himachal Pradesh.
next the Death as percentage of Confirmed is highesh in Punjab and lowest in Mizoram. 

```python
Sorted_by_Cured = merged_data.sort_values(by = 'Cured %', ascending = False)
Sorted_by_Death = merged_data.sort_values(by = 'Deaths %', ascending = False)
print(Sorted_by_Cured[['State','Cured %']].head())
print(Sorted_by_Death[['State','Deaths %']].head())
print(Sorted_by_Cured[['State','Cured %']].tail())
print(Sorted_by_Death[['State','Deaths %']].tail())
```

Now i would like to compare different statistics of different states for different months. For this i have created a fuction which takes three arguments any State/Union territory, month and the statistics which you want. 

```python
def info(Select_state,Select_month,value):
    particular_state = covid19_data[covid19_data['State'] == Select_state]
    particular_states = particular_state[particular_state['Month'] == Select_month]
    x = particular_states['Date']
    y = particular_states[value]
    plt.figure(figsize = (13,10))
    plt.xticks(rotation = 45)
    plt.plot(x,y)
    plt.xlabel(Select_month)
    plt.ylabel('No of individuals')
    plt.title(Select_state)
    return plt.show()
```

Now let us compare the death in delhi in the month of July to the month of November.

```python
info('Delhi','July','Deaths');info('Delhi','November','Deaths')
```

We notice that the number of Death in Delhi in almost 7 month have increased by approximately 1.5 times. 

Now let us compare the number of Confirmed individuals for Kerala in March to Confirmed individuals in November.

```python
info('Kerala','March','Confirmed');info('Kerala','November','Confirmed')
```

The number of Confirmed Individual in Kerala have increased exponentially. 

```python
info('Maharashtra','March','Cured');info('Maharashtra','November','Cured')
```
